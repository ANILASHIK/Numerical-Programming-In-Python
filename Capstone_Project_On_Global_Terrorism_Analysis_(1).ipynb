#### **Chart - 6.Intensity of terror attacks over various countries**

Barplot is used for there are many countries as categories.

Iraq, Pakistan & Afghanistan are the hotspots for terrotists and they bear the brunt of the terrorism the highest, righfullt so.

#### **Chart - 7.Civilian Deaths across Various Countries**

````python
# Chart - 7 visualization code
gtd_7 = df2.groupby('country_name')['deaths'].sum().reset_index().sort_values(by = 'deaths', ascending = False)
gtd_7.head()
````

````python
deaths = df2[(df2['country_name']== 'Iraq')|(df2['country_name']== 'India')|(df2['country_name']== 'Pakistan')|(df2['country_name']== 'Afghanistan')|(df2['country_name']== 'Nigeria')]
````

````python
deaths1 = deaths.groupby(['year','country_name']).agg({'deaths':'count'}).unstack()
deaths1.fillna(0, inplace = True)
deaths1.head()
````

````python
deaths1.plot(kind='area',stacked=False).legend(fontsize=20)
plt.title('Deaths Due to Terrorism Over various countries',fontsize=20)
plt.xlabel('Year',fontsize=20)
plt.ylabel('Deaths',fontsize=20)
plt.show()
````

##### 1. Why did you pick the specific chart?

Area plot is used to show the difference in number of deaths in different countries more realistic than using a barplot.

##### 2. What is/are the insight(s) found from the chart?

Irag and Afghanistan have more number of deaths.

#### **Chart - 9.Which terrorist group is carrying out the most number of attacks**

````python
# Chart - 9 visualization code
tg=df2.groupby("gname").agg({'gname':'count'}).rename(columns={"gname":"num_of_attacks"}).reset_index()
tg.sort_values(by='num_of_attacks',ascending=False,inplace=True)
tg.head()
````

The Taliban and ISIS carried out more attacks.

#### **Chart - 10.Types of Casualties**

````python
# Chart - 10 visualization code
tc = df2.groupby('year').agg({'eventid':"count",'deaths':'sum',"injured":'sum'})
tc.rename(columns={'eventid':"num_of_attacks",'deaths':"num_of_civilians_killed","injured":"num_of_civilians_wounded"},inplace=True)
tc.reset_index(inplace=True)
tc.head()
````

````python
plt.rcParams["figure.figsize"]=(25,10)
tc.plot(x="year",y=['num_of_attacks','num_of_civilians_killed','num_of_civilians_wounded'],kind="area",stacked=False).legend(fontsize=20)
plt.title("Trend in number of terrorist attacks,deaths and wounded civilians over the past five decades",fontsize=25)
plt.xlabel("Year",fontsize=20)
plt.ylabel("Count",fontsize=20)

plt.show()
````

##### 1. Why did you pick the specific chart?

Area plot is used since there are only 2 categories either wounded or dead on number of attacks. Are plot will be more pleasing.

##### 2. What is/are the insight(s) found from the chart?

The Number of civilians killed  is found to be more than number of civilians injured during attacks other than in the year 1973.

#### **Chart - 12.Pair Plot**

````python
# Chart - 12 visualization code
selected_variables = ['deaths', 'year']
selected_df = df2[selected_variables]

# Create pair plot with selected variables
sns.pairplot(selected_df, height=5,aspect=3, hue="deaths",palette="deep")
plt.show()
````

````python
selected = ['country_name', 'attacktype']
selected_df1 = df2[selected]

# Create pair plot with selected variables
sns.pairplot(selected_df1, x_vars=['attacktype'], y_vars=['country_name'],height=10, aspect=5, hue="attacktype",palette="deep")
plt.show()
````

##### 1. Why did you pick the specific chart?

Pair Plot is used because it helps to visualize the relationship between different variables and is particularly useful for identifying patterns and correlations
